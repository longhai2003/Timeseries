```python
from statsmodels.tsa.statespace.structural import UnobservedComponents
import matplotlib.pyplot as plt
```

M hnh 1 Kalman filter

```python
ts = df.sort_values("id")["Listening_Time_minutes"].reset_index(drop=True)
```

```python
model_1 = UnobservedComponents(ts, level='local level')
result_1 = model_1.fit()
```

```python
ts_filtered_1 = result_1.filtered_state[0]
```

```python
plt.figure(figsize=(12, 5))
plt.plot(ts, label='Observations', color='gray')
plt.plot(ts_filtered_1, label='Kalman Filter - Local Level', color='blue')
plt.title("Kalman Filter - Local Level Model")
plt.xlabel("Time Step (Episode ID Order)")
plt.ylabel("Listening Time (minutes)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

M hnh 2 Xu hng tuyn tnh

```python
model_2 = UnobservedComponents(ts, level='local linear trend')
result_2 = model_2.fit()
```

```python
ts_small = ts.iloc[:5000]
```

```python
model_2_small = UnobservedComponents(ts_small, level='local linear trend')
result_2_small = model_2_small.fit()
```

```python
ts_filtered_2_small = result_2_small.filtered_state[0]
```

```python
plt.figure(figsize=(10, 4))
plt.plot(ts_small, label='Observations', color='gray')
plt.plot(ts_filtered_2_small, label='Kalman Filter - Linear Trend', color='green')
plt.title("Kalman Filter - Local Linear Trend Model (5000 samples)")
plt.xlabel("Time Step")
plt.ylabel("Listening Time (minutes)")
plt.legend()
plt.tight_layout()
plt.grid(True)
plt.show()
```

M hnh 3  Kalman filter vi xu hng v thnh phn ma v

```python
from statsmodels.tsa.statespace.structural import UnobservedComponents
import matplotlib.pyplot as plt
```

```python
seasonal_period = 7
```

```python
model_seasonal = UnobservedComponents(
    ts_small,
    level='local linear trend',
    seasonal=seasonal_period
)
result_seasonal = model_seasonal.fit()
```

```python
filtered_level = result_seasonal.filtered_state[0]
```

```python
plt.figure(figsize=(10, 4))
plt.plot(ts_small, label='Observations', color='gray')
plt.plot(filtered_level, label='Filtered Level (Trend + Seasonal)', color='purple')
plt.title("Kalman Filter - Linear Trend + Seasonal (5000 samples)")
plt.xlabel("Time Step")
plt.ylabel("Listening Time (minutes)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
